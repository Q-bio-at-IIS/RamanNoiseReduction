# Validation of NRM for Raman spectral datasets

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy import io

# Plot
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["font.size"] = 16


import sys
sys.path.append("../../scripts")
from noiseReductionMethodology import preprocessing, Raman_model
```

## Load and preprocess data

```python
mat = io.loadmat("../../data/Matlab codes/data.mat")
```

```python
RAMAN_ALL = pd.DataFrame(np.vstack([mat["spombe_raman"][i][1][:, 1:-1] for i in range(len(mat["spombe_raman"]))]))
RAMAN_ALL["label"] = np.hstack([[i] * len(mat["spombe_raman"][i][1]) for i in range(len(mat["spombe_raman"]))])
RAMAN_ALL.head()
```

```python
nameList = np.array([mat["spombe_raman"][i][0][0] for i in range(len(mat["spombe_raman"]))])
conditionNum = len(nameList)
nameList
```

```python
RAMAN = RAMAN_ALL.iloc[:, 1:600].copy()
RAMAN["label"] = RAMAN_ALL["label"]
RAMAN_PROCESSED = preprocessing(RAMAN)
RAMAN_PROCESSED.head()
```

```python
RAMAN.to_csv("../../data/RAMAN_FINGERPRINT.csv", index=None)
```

## [Fig. 2b] Validate the utility of NRM with pseudo Raman spectral datasets

```python
cm_Blues = plt.get_cmap("Blues", 4)
```

```python
MEAN_RAMAN_PROCESSED = RAMAN_PROCESSED.groupby("label").mean().reset_index(drop=True)
MEAN_RAMAN_PROCESSED["label"] = range(MEAN_RAMAN_PROCESSED.shape[0])
MEAN_RAMAN_PROCESSED.head()
```

```python
numList_pseudo = np.arange(10, 205, 10)
```

```python
%%time
eigenvalues_PCA_psudodata = []
eigenvalues_NRM_psudodata = []

for n in numList_pseudo:
    print(".", end="")
    eigen_PCA_n = []
    eigen_NRM_n = []
    
    for seed in range(100):
        mean = np.zeros(MEAN_RAMAN_PROCESSED.shape[1] - 1)
        cov = np.diag(np.ones(MEAN_RAMAN_PROCESSED.shape[1] - 1))
        noise = stats.multivariate_normal(mean, cov).rvs(size=n * conditionNum, random_state=seed)
    
        OUT = pd.DataFrame([])
    
        for c in range(conditionNum):
            OUT_i = pd.DataFrame(np.vstack([[MEAN_RAMAN_PROCESSED[MEAN_RAMAN_PROCESSED["label"] == c].values[:, :-1].flatten()] for _ in range(n)]))
            OUT_i["label"] = c
            OUT = pd.concat([OUT, OUT_i], axis=0, ignore_index=True)
    
        OUT.iloc[:, :-1] += noise
    
        raman_model_ = Raman_model(OUT, cutRange=100)
        raman_model_.calcTransformation()

        eigen_PCA_n.append(raman_model_.lambda_hat[:3])
        eigen_NRM_n.append(raman_model_.lambda_tilde[:3])
        
    eigenvalues_PCA_psudodata.append(eigen_PCA_n)
    eigenvalues_NRM_psudodata.append(eigen_NRM_n)

print("")
```

```python
seed = 0
n_inf = 1000

mean = np.zeros(MEAN_RAMAN_PROCESSED.shape[1] - 1)
cov = np.diag(np.ones(MEAN_RAMAN_PROCESSED.shape[1] - 1))
noise = stats.multivariate_normal(mean, cov).rvs(size=n_inf * conditionNum, random_state=seed)
    
OUT = pd.DataFrame([])

for i in range(conditionNum):
    OUT_i = pd.DataFrame(np.vstack([[MEAN_RAMAN_PROCESSED[MEAN_RAMAN_PROCESSED["label"] == i].values[:, :-1].flatten()] for _ in range(n_inf)]))
    OUT_i["label"] = i
    OUT = pd.concat([OUT, OUT_i], axis=0, ignore_index=True)

OUT.iloc[:, :-1] += noise

raman_model_ = Raman_model(OUT, cutRange=10, cutMode="dim")
raman_model_.calcTransformation()

eigenvalues_PCA_psudodata_inf = raman_model_.lambda_hat[:3]
eigenvalues_NRM_psudodata_inf = raman_model_.lambda_tilde[:3]
```

```python
plt.figure(figsize=(6, 4))

for i in range(3):
    mean = np.array([np.mean([a[i] for a in data]) / eigenvalues_PCA_psudodata_inf[i] 
                     for data in eigenvalues_PCA_psudodata])
    std = np.array([np.std([a[i] for a in data]) / eigenvalues_PCA_psudodata_inf[i] 
                    for data in eigenvalues_PCA_psudodata])
    label = f"$\lambda_{i+1}$" + "$^\mathrm{PCA}$"
    plt.plot(numList_pseudo, mean, marker="o", color=cm_Blues(i + 1), linestyle="dashed", clip_on=False,
             label=label, zorder=10, linewidth=2)
    
for i in range(3):
    mean = np.array([np.mean([a[i] for a in data]) / eigenvalues_PCA_psudodata_inf[i] 
                     for data in eigenvalues_NRM_psudodata])
    std = np.array([np.std([a[i] for a in data]) / eigenvalues_PCA_psudodata_inf[i] 
                    for data in eigenvalues_NRM_psudodata])
    label = f"$\lambda_{i+1}$" + "$^\mathrm{NRM}$"
    plt.plot(numList_pseudo, mean, marker="o", color=cm_Blues(i + 1), clip_on=False, zorder=10,
             label=label, linewidth=2)

plt.hlines(1, 0, 300, color="k", linewidth=1, linestyle="dashed")
plt.xlim(0, 200)
plt.xticks(range(0, 210, 50))
plt.yticks(np.hstack([1, np.arange(0, 9.1, 3)]))
plt.ylim(0, 9)
plt.xlabel("Sample size per condition")
plt.ylabel("Ratio")
plt.legend(ncol=2)
plt.tight_layout()
plt.show()
```

## [Fig. 2d and e] Validate the NRM’s effectiveness for the actual Raman spectral dataset

```python
GROUP = [RAMAN_PROCESSED[RAMAN_PROCESSED["label"] == i] for i in range(conditionNum)]
numList = np.arange(6, 60, 6)
numList
```

```python
raman_model_all = Raman_model(RAMAN_PROCESSED, cutRange=100)
raman_model_all.calcTransformation()
```

```python
eigenvalues_PCA_list = []
eigenvalues_NRM_list = []

for num in numList:
    out = []
    out_NRM = []
    np.random.seed(0)
    for i in range(100):
        INPUTDATA = pd.DataFrame([])
        for OUT in GROUP:
            idxList = list(np.arange(OUT.shape[0]))
            np.random.shuffle(idxList)
            INPUTDATA = pd.concat([INPUTDATA, OUT.iloc[idxList[:num], :]], axis=0)
        INPUTDATA = INPUTDATA.reset_index(drop=True)

        raman_model_ = Raman_model(INPUTDATA, cutRange=100)
        raman_model_.calcTransformation()
        out.append(raman_model_.lambda_hat)
        out_NRM.append(raman_model_.lambda_tilde)
        
    eigenvalues_PCA_list.append(out)
    eigenvalues_NRM_list.append(out_NRM)
```

```python
plt.figure(figsize=(6, 4))


for dataList, name, linestyle in zip([eigenvalues_PCA_list, eigenvalues_NRM_list], 
                                     ["PCA", "NRM"], ["--", "--"]):
    for i in range(3):
        out = [np.mean(data, axis=0)[i] / raman_model_all.lambda_hat[i] 
               for data in dataList]
        
        if name == "PCA":
            label = f"$\lambda_{i+1}$" + "$^\mathrm{PCA}$"
        else:
            label = f"$\lambda_{i+1}$" + "$^\mathrm{NRM}$"
            
        plt.plot(numList, out, marker="o", 
                 color=cm_Blues(i + 1), linestyle=linestyle, clip_on=False, zorder=10, linewidth=2, 
                 label=label)
    
plt.hlines(1, 0, 100, color="k", linewidth=1, linestyle="dashed")

plt.xticks(numList)
plt.xlim(numList[0] - 2, numList[-1] + 2)
plt.yticks(np.arange(0, 1.3, 0.2))
plt.ylim(0.6, 1.2)
plt.xlabel("Sample size per condition")
plt.ylabel("Ratio")
plt.legend(ncol=2)
plt.tight_layout()
plt.show()
```

```python
%%time

k = 20

pointDiff_PCA_list = []
pointDiff_NRM_list = []

for num in numList:
    print(".", end="")
    out = []
    out_NRM = []
    np.random.seed(0)
    for i in range(100):
        INPUTDATA = pd.DataFrame([])
        for OUT in GROUP:
            idxList = list(np.arange(OUT.shape[0]))
            np.random.shuffle(idxList)
            INPUTDATA = pd.concat([INPUTDATA, OUT.iloc[idxList[:num], :]], axis=0)
        BASE = raman_model_all.RAMAN_PCA.iloc[INPUTDATA.index, :].reset_index(drop=True)
        INPUTDATA = INPUTDATA.reset_index(drop=True)

        raman_model_ = Raman_model(INPUTDATA, cutMode="dim", cutRange=k)
        raman_model_.calcTransformation()
        
        rotation = np.array([np.dot(a, b) for a, b in zip(raman_model_all.H_hat.T, raman_model_.H_hat.T)])
        BASE_mod = BASE.copy()
        BASE_mod.loc[:, np.where(rotation < 0)[0]] *= -1
        BASE_mod = BASE_mod.loc[:, range(k)]
        BASE_mod["label"] = BASE["label"].values

        
        out.append(np.mean((raman_model_.RAMAN_PCA.values[:, :-1] - BASE_mod.values[:, :-1]) ** 2, axis=0))
        out_NRM.append(np.mean((raman_model_.RAMAN_NRM.values[:, :-1] - BASE_mod.values[:, :-1]) ** 2, axis=0))

    pointDiff_PCA_list.append(out)
    pointDiff_NRM_list.append(out_NRM)
    
print("")
```

```python
plt.figure(figsize=(6, 4))

diff_mean = np.vstack([np.mean(np.array(out_nrm) - np.array(out), axis=0)
                       for out, out_nrm in zip(pointDiff_PCA_list, pointDiff_NRM_list)])
diff_std = np.vstack([np.std(np.array(out_nrm) - np.array(out), axis=0)
                      for out, out_nrm in zip(pointDiff_PCA_list, pointDiff_NRM_list)])

for i in range(3):
    plt.plot(numList, diff_mean[:, i], color=cm_Blues(i + 1),
             clip_on=False, marker="o", zorder=10, label=f"Axis {i + 1}", linewidth=2)
    
plt.hlines(0, 0, 100, color="k", linewidth=1, linestyle="dashed", zorder=0)

plt.xticks(numList)
plt.xlim(numList[0] - 2, numList[-1] + 2)
plt.yticks(np.arange(-0.015, 0.0051, 0.005))
plt.ylim(-0.015, 0.005)
plt.legend(loc="lower right")
plt.xlabel("Sample size per condition")
plt.ylabel("Diff of MSE")
plt.tight_layout()
plt.show()
```
