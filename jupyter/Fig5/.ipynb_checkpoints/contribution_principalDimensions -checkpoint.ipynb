## [Fig. 5d] Contribution of principal dimensions on transcriptomic profile estimation

```python
SUMMARY_CLASSIFICATION = pd.read_csv("../../results/classification/LDA/ACCLIST_ALLDATA_LDA_PERCENT.csv", index_col=0)
maxPercent = float(list(SUMMARY_CLASSIFICATION.sort_values("PCA", ascending=False).index)[0].replace("%", ""))
maxPercent
```

```python
raman_model_all = Raman_model(RAMAN_PROCESSED, cutMode="percent", cutRange=maxPercent)
raman_model_all.calcTransformation()
maxDim = raman_model_all.k_hat
maxDim
```

```python
SUMMARY_ESTIMATION = pd.read_csv(f"../../results/SUMMARY_PERCENT_fixedDim_SPOMBE.csv", index_col=0)
press_base = SUMMARY_ESTIMATION.loc[[f"{maxPercent:.1f}%"]]["PRESS_PCA"].values[0]
press_base
```

```python
%%time
predict_list_shuffled_maxAcc = []

loopNum = 100

for i in range(1, maxDim + 1):
    num = special.comb(maxDim, i, exact=True)
    print(".", end="")
    if num < loopNum:
        shuffledIdx = [a for a in itertools.combinations(range(maxDim), i)]
    else:
        shuffledIdx = [np.random.choice(np.arange(maxDim), size=i) for _ in range(loopNum)]
        
    out = []
    for pickedIdx in shuffledIdx:
        pickedIdx = np.hstack([pickedIdx, -1])
        
        lda_model = LDA_model()
        DATA_LDA = lda_model.fit_transform(raman_model_all.RAMAN_PCA.iloc[:, pickedIdx])

        DATA_LDA = DATA_LDA.groupby("label").mean()
        DATA_LDA["label"] = np.arange(DATA_LDA.shape[0])

        PREDICT = calcPrediction(TRANSCRIPTOME, DATA_LDA, n_components=0, max_iter=50000)

        out.append(np.sum((returnValues(PREDICT) - returnValues(TRANSCRIPTOME)) ** 2, axis=1).sum())

    predict_list_shuffled_maxAcc.append(out)

print("")
```

```python
predict_array_shuffled_maxAcc = np.ones((len(predict_list_shuffled_maxAcc), loopNum)) * np.nan

for i, data in enumerate(predict_list_shuffled_maxAcc):
    predict_array_shuffled_maxAcc[i, :len(data)] = data

np.save("../../results/predict_array_shuffled_maxAcc.npy", predict_array_shuffled_maxAcc)
```

```python
fig = plt.figure(figsize=(6, 4))

for i, out in enumerate(predict_array_shuffled_maxAcc[:-1]):
    if i <= 8:
        continue
    else:
        plt.scatter(np.ones(len(out)) * (i + 1), out, color="#00000000", edgecolor="#5D999B", marker=".")

plt.hlines(press_maxParam, 0, 70, color="k", linewidth=1, zorder=0, linestyle="dashed")

plt.xticks(np.arange(0, 71, 10))
plt.xlim(0, 70)
plt.yscale("log")
plt.yticks([10 ** i for i in [11, 13, 15]])
plt.ylim(1e11, 1e15)
plt.xlabel("Dimension")
plt.ylabel("PRESS")
plt.minorticks_off()
plt.tight_layout()
plt.show()
```

## [Fig. 5e] Contribution of principal dimensions on classification

```python
acc_base = np.max(SUMMARY_CLASSIFICATION["PCA"].values)
acc_base
```

```python
accList = []

for i in range(1, maxDim + 1):
    lda_model_all = LDA_model()
    pickedIdx = np.hstack([np.arange(i), -1])
    accList_ = lda_model_all.clustering_LDA(raman_model_all.RAMAN_PCA.iloc[:, pickedIdx], isPrint=False)

    accList.append(accList_)
```

```python
%%time
classification_list_shuffled = []

loopNum = 100

for i in range(1, maxDim + 1):    
    num = special.comb(maxDim, i, exact=True)
    print(".", end="")
    if num < loopNum:
        shuffledIdx = [a for a in itertools.combinations(range(raman_model_all.k_hat), i)]
    else:
        shuffledIdx = [np.random.choice(np.arange(raman_model_all.k_hat), size=i) for _ in range(loopNum)]
        
    out = []
    for pickedIdx in shuffledIdx:
        pickedIdx = np.hstack([pickedIdx, -1])
        
        
        lda_model_all = LDA_model()
        accList_ = lda_model_all.clustering_LDA(raman_model_all.RAMAN_PCA.iloc[:, pickedIdx], isPrint=False)
        
        out.append(np.mean(accList_))

    classification_list_shuffled.append(out)

print("")
```

```python
classification_array_shuffled = np.ones((len(classification_list_shuffled), loopNum)) * np.nan

for i, data in enumerate(classification_list_shuffled):
    classification_array_shuffled[i, :len(data)] = data
    
np.save("../../results/classification_array_shuffled_maxAcc.npy", classification_array_shuffled)
```

```python
fig = plt.figure(figsize=(6, 4))

plt.hlines(acc_base, 0, 80, color="k", linewidth=1, zorder=0, linestyle="dashed")

mean = np.mean(accList, axis=1)
plt.plot(np.arange(len(mean)) + 1, mean, color="k", linewidth=2, zorder=0)

for i, out in enumerate(classification_array_shuffled[:-1]):
    plt.scatter(np.ones(len(out)) * (i + 1), out, color="#00000000", edgecolor="#89AF82", marker=".")


plt.xticks(np.arange(0, 71, 10))
plt.xlim(0, 70)
plt.xlabel("Dimension")
plt.yticks(np.arange(0, 101, 20))
plt.ylim(0, 100)
plt.ylabel("Accuracy (%)")
plt.tight_layout()
plt.show()
```

## [Fig. 5f] Comparision of two tasks: classification and transcriptomic profile estimation

```python
plt.figure(figsize=(6, 4))
    
out = [(out[~np.isnan(out)] <= press_base).sum() / len(out) for out in predict_array_shuffled_maxAcc]
out = out[1:-2]
plt.plot(np.arange(len(out)) + 2, out, color="#5D999B", clip_on=False, zorder=20, linewidth=2, label="Estimation")

out = [(out[~np.isnan(out)] >= acc_base).sum() / len(out) for out in classification_array_shuffled]
out = out[1:-2]
plt.plot(np.arange(len(out)) + 2, out, color="#89AF82", clip_on=False, zorder=20, linewidth=2, label="Classification")

plt.hlines([0.05, 0.01], 0, 70, color="k", linewidth=1, linestyle="dashed", zorder=0)

plt.xticks(np.arange(0, 61, 10))
plt.xlim(0, len(out) + 1)
plt.xlabel("Dimension")
plt.yticks(np.arange(0, 0.21, 0.05))
plt.ylim(0, 0.2)
plt.ylabel("Ratio")

plt.legend()
plt.tight_layout()
plt.show()
```
