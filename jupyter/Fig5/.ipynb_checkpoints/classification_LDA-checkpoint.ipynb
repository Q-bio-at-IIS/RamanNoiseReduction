## Classification analysis

### All samples

````python
RAMAN = pd.read_csv("../../data/RAMAN_FINGERPRINT.csv")
RAMAN_PROCESSED = preprocessing(RAMAN)
GROUP = [RAMAN_PROCESSED[RAMAN_PROCESSED["label"] == i].reset_index(drop=True) for i in range(RAMAN_PROCESSED["label"].max() + 1)]
````

````python
cutRanges = np.arange(60, 100, 0.5)
cutRangesStr = [f"{a:.0f}" if a % 1 == 0 else f"{a:.1f}".replace(".", "")
                for a in cutRanges]
````

````python
accList_pca = []
accList_nrm = []

dimList_pca = []
dimList_nrm = []

for i in cutRanges:
    raman_model_ = Raman_model(RAMAN_PROCESSED, cutMode="percent", cutRange=i)
    raman_model_.calcTransformation()

    dimList_pca.append(raman_model_.k_hat)
    dimList_nrm.append(raman_model_.k_tilde)
    accList_pca.append(classification_method(raman_model_.RAMAN_PCA, method="LDA", isPrint=False))
    accList_nrm.append(classification_method(raman_model_.RAMAN_NRM, method="LDA", isPrint=False))
    
    print(".", end="")
print("")
````

````python
RESULTS = pd.DataFrame(np.vstack([np.mean(accList_pca, axis=1), np.mean(accList_nrm, axis=1), dimList_pca, dimList_nrm]).T,
                       columns=["PCA", "NRM", "dim_PCA", "dim_NRM"],
                       index=[f"{a:.1f}%" for a in cutRanges])

RESULTS.to_csv("../../results/classification/LDA/ACCLIST_ALLDATA_LDA_PERCENT.csv", index=True)
RESULTS.head()
````

## [Fig. 5a] Illustration of LDA space

````python
def makeEllipse(x, y, n_sigma=2, color="k"):
    cov = np.cov(x, y) 
    eigenvalues, eigenvectors = np.linalg.eig(cov)

    w, h = 2 * n_sigma * np.sqrt(eigenvalues)
    theta_rad = np.arctan2(*eigenvectors[:,0][::-1])
    theta = np.degrees(theta_rad)

    ell = patches.Ellipse(xy=(np.mean(x), np.mean(y)), width=w, height=h, angle=theta, color=color, linewidth=2)
    ell.set_facecolor('none')
    return ell
````

````python
raman_model_all = Raman_model(RAMAN_PROCESSED, cutMode="percent_fixedDim", cutRange=96.5)
raman_model_all.calcTransformation()
lda_model_all = LDA_model()
ALL_LDA = lda_model_all.fit_transform(raman_model_all.RAMAN_PCA)
````

````python
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot()

for i, color in enumerate(condition_color):
    TRAIN = ALL_LDA[ALL_LDA["label"] == i]
    plt.scatter(TRAIN.values[:, 0], TRAIN.values[:, 1], color=color, s=10, marker=".", alpha=1)
    ell = makeEllipse(TRAIN.values[:, 0], TRAIN.values[:, 1], color=color)
    ax.add_artist(ell)

plt.xticks(np.arange(-1.0, 2.1, 1))
plt.yticks(np.arange(-1.2, 1.3, 1.2))
plt.xlim(-1, 2)
plt.ylim(-1.2, 1.2)
plt.xlabel(f"LD1")
plt.ylabel(f"LD2")

plt.tight_layout()
plt.show()
````

## [Fig. 5b and SFig. 2a] Classification accuracy at the threshold optimized for PCA-based transcriptomic profile estimation

````python
SUMMARY_PERCENTAGE_fixedDim = pd.DataFrame([])

for n in nListStr:
    if n == "All":
        OUT= pd.read_csv("../../results/SUMMARY_PERCENT_fixedDim_SPOMBE.csv", index_col=0).sort_values(["PRESS_PCA"]).reset_index(drop=False).loc[[0], :]

    else:
        OUT = pd.read_csv(f"../../results/SUMMARY_PERCENT_fixedDim_SPOMBE_dataSize{n}.csv", index_col=0).sort_values(["PRESS_PCA"]).reset_index(drop=False).loc[[0], :]
    
    OUT["sampleSize"] = n
    SUMMARY_PERCENTAGE_fixedDim = pd.concat([SUMMARY_PERCENTAGE_fixedDim, OUT], axis=0, ignore_index=True)
    
SUMMARY_PERCENTAGE_fixedDim["percent_PCA"] = [float(a.replace("%", "")) for a in SUMMARY_PERCENTAGE_fixedDim["index"].values]
SUMMARY_PERCENTAGE_fixedDim.head()
````

````python
dimList_nrm = []
dimList_pca = []

for (per, n) in SUMMARY_PERCENTAGE_fixedDim.loc[:, ["index", "sampleSize"]].values:
    print(".", end="")
    if n == "All":
        OUT = pd.read_csv("../../results/classification/LDA/ACCLIST_ALLDATA_LDA_PERCENT.csv", index_col=0)
        dimList_pca.append(OUT.loc[[per]]["dim_PCA"].values[0])
        dimList_nrm.append(OUT.loc[[per]]["dim_NRM"].values[0])
        
    else:
        np.random.seed(0)
        per = float(per.replace("%", ""))
        n = int(n)
        out_nrm = []
        out_pca = []
        
        for j in range(100):
            INPUTDATA = pd.DataFrame([])
            for OUT in GROUP:
                idxList = list(np.arange(OUT.shape[0]))
                np.random.shuffle(idxList)
                INPUTDATA = pd.concat([INPUTDATA, OUT.iloc[idxList[:n], :]], axis=0)
            INPUTDATA = INPUTDATA.reset_index(drop=True)
            
            raman_model_ = Raman_model(INPUTDATA, cutMode="percent", cutRange=per)
            raman_model_.calcTransformation()

            out_pca.append(raman_model_.k_hat)
            out_nrm.append(raman_model_.k_tilde)
    
        dimList_nrm.append(np.mean(out_nrm))
        dimList_pca.append(np.mean(out_pca))
````

````python
plt.figure(figsize=(6, 4))

for data, color, name in zip([dimList_nrm, dimList_pca], colorList, nameList):
    plt.plot(data, color=color, marker="o", label=name)
    
plt.xticks(np.arange(len(nListStr)), nListStr)
plt.xlabel("Sample size per condition")
plt.yticks(np.arange(0, 151, 50))
plt.ylabel("Number of dimensions")
plt.legend()
plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(6, 4))

for i, (per, n) in enumerate(SUMMARY_PERCENTAGE_fixedDim.loc[:, ["index", "sampleSize"]].values):
    if n == "All":
        OUT = pd.read_csv("../../results/classification/LDA/ACCLIST_ALLDATA_LDA_PERCENT.csv", index_col=0)
        out_pca, out_nrm = OUT.loc[per, ["PCA", "NRM"]]
        
    else:
        per = float(per.replace("%", ""))
        if per % 1 == 0:
            per = str(int(per))
        else:
            per = f"{per:.1f}".replace(".", "")
    
        out_pca = pd.read_csv(f"../../results/classification/LDA/ACCLIST_NUMEFFECTS_PCA_cutPercent{per}.csv")[f"n={n}"].values
        out_nrm = pd.read_csv(f"../../results/classification/LDA/ACCLIST_NUMEFFECTS_NRM_cutPercent{per}.csv")[f"n={n}"].values
        
    p = stats.wilcoxon(out_pca, out_nrm, alternative="less").pvalue
    text = ""
    if p < 0.001:
        text = "***"
    elif p < 0.01:
        text = "**"
    elif p < 0.05:
        text = "*"

    for j, (data, color) in enumerate(zip([out_pca, out_nrm], colorList)):
        mean = np.mean(data)
        std = np.std(data)
        plt.bar(i + (j - 0.5) * 0.4, mean, color=color, width=0.4, yerr=std, ecolor=color)

    plt.text(i, np.mean(out_nrm) + 10, text, ha="center", va="center", fontsize=10)
        
plt.xticks(range(len(nListStr)), nListStr)
plt.xlabel("Sample size per condition")
plt.yticks(np.arange(0, 110, 20))
plt.ylim(0, 100)
plt.ylabel("Accuracy (%)")

patchList = [patches.Patch(color=color, label=name) for color, name in zip(colorList, nameList)]
plt.legend(handles=patchList, ncol=2, loc="lower left")
plt.tight_layout()
plt.show()
````

## [SFig. 2b] Transcriptomic profile estimation with the same threshold for explained variance ratio

````python
TRANSCRIPTOME = pd.read_csv("../../data/TRANSCRIPTOME.csv")
````

````python
def generateDataList(n, GROUP=GROUP):
    RAMAN_dataList = []
    np.random.seed(0)
    for j in range(100):
        INPUTDATA = pd.DataFrame([])
        for OUT in GROUP:
            idxList = list(np.arange(OUT.shape[0]))
            np.random.shuffle(idxList)
            INPUTDATA = pd.concat([INPUTDATA, OUT.iloc[idxList[:n], :]], axis=0)
        INPUTDATA = INPUTDATA.reset_index(drop=True)

        RAMAN_dataList.append(INPUTDATA)
    return RAMAN_dataList
````

````python
def performTranscriptomicProfileEstimation(dataList, cutPercentages):
    predict_dataList = []
    percent_to_dim_PCA_dataList = []
    percent_NRM_dataList = []

    for RAMAN_DATA in dataList:
        print(".", end="")
        predict_list = []
        percent_to_dim_PCA = []
        percent_NRM = []

        for cutPercent in cutPercentages:
            raman_model = Raman_model(RAMAN_DATA, cutRange=cutPercent, cutMode="percent")
            raman_model.calcTransformation()

            out = []
            for DATA in [raman_model.RAMAN_PCA, raman_model.RAMAN_NRM]:
                lda_model = LDA_model()
                DATA_LDA = lda_model.fit_transform(DATA)

                DATA_LDA = DATA_LDA.groupby("label").mean()
                DATA_LDA["label"] = np.arange(DATA_LDA.shape[0])

                PREDICT = calcPrediction(TRANSCRIPTOME, DATA_LDA, n_components=0, max_iter=50000)

                out.append(PREDICT)

            predict_list.append(out)
            percent_to_dim_PCA.append(raman_model.k_hat)
            percent_NRM.append(raman_model.percent_tilde)

        predict_dataList.append(predict_list)
        percent_to_dim_PCA_dataList.append(percent_to_dim_PCA)
        percent_NRM_dataList.append(percent_NRM)
    print("")
    return predict_dataList, percent_to_dim_PCA_dataList, percent_NRM_dataList
````

````python
diffList = []
pList = []

for (per, n) in SUMMARY_PERCENTAGE_fixedDim.loc[:, ["index", "sampleSize"]].values:
    print(f"[{n}]", end="")
    per = float(per.replace("%", ""))
    
    if n == "All":
        predict_dataList, percent_to_dim_PCA_dataList, percent_NRM_dataList = performTranscriptomicProfileEstimation([RAMAN_PROCESSED], [per])
        
    else:
        dataList = generateDataList(int(n))
        predict_dataList, percent_to_dim_PCA_dataList, percent_NRM_dataList = performTranscriptomicProfileEstimation(dataList, [per])

    out_pca = np.vstack([np.array([np.sum((returnValues(data[0]) - returnValues(TRANSCRIPTOME)) ** 2, axis=1).sum()
                               for data in predict_list])
                     for predict_list in predict_dataList])

    out_nrm = np.vstack([np.array([np.sum((returnValues(data[1]) - returnValues(TRANSCRIPTOME)) ** 2, axis=1).sum()
                                   for data in predict_list])
                         for predict_list in predict_dataList])

    if n == "All":
        p = np.nan
    else:
        p = np.array([stats.wilcoxon(a, b, alternative="two-sided").pvalue for a, b in zip(out_pca.T, out_nrm.T)])[0]

    diffList.append(out_nrm - out_pca)
    pList.append(p)
````

````python
plt.figure(figsize=(6, 4))

diff_mean = [np.mean(data) for data in diffList]
diff_std = [np.std(data) / np.sqrt(data.shape[0]) for data in diffList]

plt.errorbar(range(len(diff_mean)), diff_mean, yerr=diff_std, marker="o", color="#396C75",
             linewidth=2)

for i, (p, y, y_) in enumerate(zip(pList, diff_mean, diff_std)):
    text = ""
    if p < 0.001:
        text = "***"
    elif p < 0.01:
        text = "**"
    elif p < 0.05:
        text = "*"
        
    plt.text(i, y + y_, text, ha="center", va="bottom")

plt.hlines(0, -1, len(nListStr), color="k", linewidth=1, zorder=0, linestyle="dashed")

plt.xticks(range(len(nListStr)), nListStr)
plt.xlim(-0.4, len(nListStr) - 1 + 0.4)
plt.yticks(np.arange(-1.2, 1.3, 0.6) * 1e11)
plt.ylim(-1.2e11, 1.2e11)
plt.ylabel("PRESS differences")
plt.xlabel("Sample size per condition")
plt.tight_layout()
plt.show()
````
