# Evaluation of NRM's superiority for transcriptomic profile estimation

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy import io

# Plot
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["font.size"] = 16

from matplotlib import colors as pltcolors
from matplotlib.colors import ListedColormap
```

```python
nListStr = np.hstack(["All", np.arange(12, 55, 6)[::-1]]).astype(str)
colorList = ["gray", "#B51700"]
nameList = ["PCA", "NRM"]

condition_color = np.hstack([
    [pltcolors.to_hex(plt.get_cmap("autumn", 4)(i + 1)) for i in range(2)],
    [pltcolors.to_hex("C8")],
    [pltcolors.to_hex(plt.get_cmap("summer", 3)(1 - i)) for i in range(2)],
    [pltcolors.to_hex(plt.get_cmap("Blues", 3)(i + 1)) for i in range(1)],
    [pltcolors.to_hex(a) for a in ["royalblue", "blueviolet", "violet", "sienna"]],
])


ListedColormap(condition_color)
```

```python
SUMMARY_PERCENTAGE_fixedDim = pd.DataFrame([])

for n in nListStr:
    if n == "All":
        OUT= pd.read_csv("../../results/SUMMARY_PERCENT_fixedDim_SPOMBE.csv", index_col=0).sort_values(["PRESS_PCA"]).reset_index(drop=False).loc[[0], :]

    else:
        OUT = pd.read_csv(f"../../results/SUMMARY_PERCENT_fixedDim_SPOMBE_dataSize{n}.csv", index_col=0).sort_values(["PRESS_PCA"]).reset_index(drop=False).loc[[0], :]
    
    OUT["sampleSize"] = n
    SUMMARY_PERCENTAGE_fixedDim = pd.concat([SUMMARY_PERCENTAGE_fixedDim, OUT], axis=0, ignore_index=True)
    
SUMMARY_PERCENTAGE_fixedDim["percent_PCA"] = [float(a.replace("%", "")) for a in SUMMARY_PERCENTAGE_fixedDim["index"].values]
SUMMARY_PERCENTAGE_fixedDim.head()
```

## [Fig. 4a - c] Evaluation of NRM's effects on transcriptomic profile estimation with various sample size datasets

```python
plt.figure(figsize=(6, 4))


for name, color in zip(nameList, colorList):
    plt.plot(SUMMARY_PERCENTAGE_fixedDim[f"percent_{name}"],
             color=color, label=name,
             marker="o", linewidth=2)

plt.xticks(range(len(nListStr)), nListStr)
plt.xlim(-0.4, len(nListStr) - 1 + 0.4)
plt.yticks(np.arange(96, 101, 1))
plt.ylabel("Explained variance (%)")
plt.xlabel("Sample size per condition")
plt.legend(ncol=2, loc="upper left")
plt.tight_layout()
plt.show()
```

```python
diff_mean = SUMMARY_PERCENTAGE_fixedDim["PRESS_diff"].values
diff_pval = SUMMARY_PERCENTAGE_fixedDim["PRESS_pVal"].values
diff_std = SUMMARY_PERCENTAGE_fixedDim["PRESS_diff_std"].values / np.sqrt(100)

plt.figure(figsize=(6, 4))
plt.errorbar(range(len(diff_mean)), diff_mean, yerr=diff_std, marker="o", color="#396C75",
             linewidth=2)


for i, (p, y, y_) in enumerate(zip(diff_pval, diff_mean, diff_std)):
    p *= (len(diff_mean) - 1)
    text = ""
    if p < 0.001:
        text = "***"
    elif p < 0.01:
        text = "**"
    elif p < 0.05:
        text = "*"
        
    plt.text(i, y + y_, text, ha="center", va="bottom")
    
plt.hlines(0, -1, len(nListStr), color="k", linewidth=1, zorder=0, linestyle="dashed")

plt.xticks(range(len(nListStr)), nListStr)
plt.xlim(-0.4, len(nListStr) - 1 + 0.4)
plt.yticks(np.arange(-8, 9, 4) * 1e9)
plt.ylim(-8e9, 8e9)
plt.ylabel("PRESS differences")
plt.ylabel("Sample size per condition")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))

out = SUMMARY_PERCENTAGE_fixedDim.loc[:, [f"PRESS_NRM_c{i + 1}" for i in range(10)]].values - SUMMARY_PERCENTAGE_fixedDim.loc[:, [f"PRESS_PCA_c{i + 1}" for i in range(len(condition_color))]].values
for i, (data, c) in enumerate(zip(out.T, condition_color)):
    plt.plot(data, marker="o", color=c, linewidth=2, label=i + 1)

plt.hlines(0, -1, len(nListStr), color="k", linewidth=1, zorder=0, linestyle="dashed")

plt.xticks(range(len(nListStr)), nListStr)
plt.xlim(-0.4, len(nListStr) - 1 + 0.4)
plt.yticks(np.arange(-8, 9, 4) * 1e9)
plt.ylim(-8e9, 8e9)
plt.xlabel("sample size per condition")
plt.ylabel("PRESS difference")
plt.legend(ncol=2, loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

## [SFig. 1c] Evaluation by cosine similarity

```python
diff_mean = SUMMARY_PERCENTAGE_fixedDim["cosine_diff"].values
diff_pval = SUMMARY_PERCENTAGE_fixedDim["cosine_pVal"].values
diff_std = SUMMARY_PERCENTAGE_fixedDim["cosine_diff_std"].values / np.sqrt(100)

plt.figure(figsize=(6, 4))
plt.errorbar(range(len(diff_mean)), diff_mean, yerr=diff_std, marker="o", color="#396C75",
             linewidth=2)


for i, (p, y, y_) in enumerate(zip(diff_pval, diff_mean, diff_std)):
    p *= (len(diff_mean) - 1)
    text = ""
    if p < 0.001:
        text = "***"
    elif p < 0.01:
        text = "**"
    elif p < 0.05:
        text = "*"
        
    plt.text(i, y + y_, text, ha="center", va="bottom")
    
plt.hlines(0, -1, len(nListStr), color="k", linewidth=1, zorder=0, linestyle="dashed")

plt.xticks(range(len(nListStr)), nListStr)
plt.xlim(-0.4, len(nListStr) - 1 + 0.4)
plt.yticks(np.arange(-2, 2.1, 1) * 1e-4)
plt.ylim(-2*1e-4, 2e-4)
plt.ylabel("PRESS differences")
plt.xlabel("Sample size per condition")
plt.tight_layout()
plt.show()
```
